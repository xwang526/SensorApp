```python
Pitching_front_to_back = np.array([mean_pitching_front2, mean_pitching_front3, 
                                   mean_pitching_front4, mean_pitching_front5]).flatten().tolist()
Pitching_back_to_front = np.array([mean_pitching_back2, mean_pitching_back3, 
                                   mean_pitching_back4, mean_pitching_back5]).flatten().tolist()

Rolling_left_to_right = np.array([mean_rolling_left2, mean_rolling_left3, 
                                  mean_rolling_left4, mean_rolling_left5]).flatten().tolist()
Rolling_right_to_left = np.array([mean_rolling_right2, mean_rolling_right3, 
                                  mean_rolling_right4, mean_rolling_right5]).flatten().tolist()

Yawing_left_to_right = np.array([mean_yawing_left2, mean_yawing_left3, mean_yawing_left4, 
                                 mean_yawing_left5]).flatten().tolist()
Yawing_right_to_left = np.array([mean_yawing_right2, mean_yawing_right3, mean_yawing_right4, 
                                 mean_yawing_right5]).flatten().tolist()
```

```python
# Grouped by user size
Group_of_two = np.array([mean_pitching_front2, mean_pitching_back2, mean_rolling_left2, 
                mean_rolling_right2, mean_yawing_left2, mean_yawing_right2]).flatten().tolist()

Group_of_three = np.array([mean_pitching_front3, mean_pitching_back3, mean_rolling_left3, 
                mean_rolling_right3, mean_yawing_left3, mean_yawing_right3]).flatten().tolist()

Group_of_four = np.array([mean_pitching_front4, mean_pitching_back4, mean_rolling_left4, 
                mean_rolling_right4, mean_yawing_left4, mean_yawing_right4]).flatten().tolist()

Group_of_five = np.array([mean_pitching_front5, mean_pitching_back5, mean_rolling_left5, 
                mean_rolling_right5, mean_yawing_left5, mean_yawing_right5]).flatten().tolist()
```

```python
# stack histogram of differnt user group size
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_NOTEBOOK
```

```python
# stack histogram of mean accuracy of differnt movement
bar = Bar()
bar.add_xaxis(["PitchFrontBack", "PitchBackFront", "RollLeftRight", "RollRightLeft", 
               "YawLeftRight", "YawRightLeft"])
bar.add_yaxis("Accuracy of 2 users", Group_of_two)
bar.add_yaxis("Accuracy of 3 users", Group_of_three)
bar.add_yaxis("Accuracy of 4 users", Group_of_four)
bar.add_yaxis("Accuracy of 5 users", Group_of_five)

bar.render_notebook()
```

```python
# Grouped by differnt user size
bar2 = Bar()
bar2.add_xaxis(["2 users", "3 users", "4 users", "5 users"])
bar2.add_yaxis("Pitch from front to back", Pitching_front_to_back)
bar2.add_yaxis("Pitch from back to front", Pitching_back_to_front)
bar2.add_yaxis("Roll from left to right", Rolling_left_to_right)
bar2.add_yaxis("Roll from right to left", Rolling_right_to_left)
bar2.add_yaxis("Yaw from left to right", Yawing_left_to_right)
bar2.add_yaxis("Yaw from right to left", Yawing_right_to_left)

bar2.render_notebook()
```

```python
from pyecharts.charts import Line
line_two_users = Line()
combinations = np.linspace(1,45,45,endpoint = True)
line_two_users.add_xaxis(combinations)
line_two_users.add_yaxis(series_name = "Pitch front to back", y_axis = pitching_front2, 
                         label_opts = opts.LabelOpts(is_show = False), is_smooth = True)
line_two_users.add_yaxis(series_name = "Pitch back to front", y_axis = pitching_back2, 
                         label_opts = opts.LabelOpts(is_show = False), is_smooth = True)
line_two_users.add_yaxis(series_name = "Roll left to right", y_axis = rolling_left2, 
                         label_opts = opts.LabelOpts(is_show = False), is_smooth = True)
# line_two_users.add_yaxis(series_name = "Roll right to left", y_axis = rolling_right2, 
#                          label_opts = opts.LabelOpts(is_show = False), is_smooth = True)
line_two_users.add_yaxis(series_name = "Yaw left to right", y_axis = yawing_left2, 
                         label_opts = opts.LabelOpts(is_show = False), is_smooth = True)
line_two_users.add_yaxis(series_name = "Yaw right to left", y_axis = yawing_right2, 
                         label_opts = opts.LabelOpts(is_show = False), is_smooth = True)
line_two_users.set_global_opts(
#     title_opts = opts.TitleOpts(title = "Leave-One-Out Cross Validation for each motion of two users"),
    tooltip_opts = opts.TooltipOpts(trigger = 'axis'),
    yaxis_opts = opts.AxisOpts(type_='value',axistick_opts=opts.AxisTickOpts(is_show=True),
                               name = 'Accuracy', max_interval = 0.05,
                              splitline_opts = opts.SplitLineOpts(is_show=True)),
    xaxis_opts = opts.AxisOpts(type_='category',boundary_gap = False),
)
line_two_users.render_notebook()
```
